```python
# melt function convert wide dataFrame to long dataFrame 
# Melt function convert wide dataFrame to long dataframe

pd.DataFrame({'cse':[20]}).melt()
```

```python
pd.DataFrame({'cse':[20], 'ese':[50], 'mech':[30]}).melt(var_name='branch', value_name='nu_of_students')
```

```python
pd.DataFrame({
    'branch':['cse', 'civil', 'mech'],
    '2019': [60, 70, 30],
    '2020':[70, 30, 50],
    '2021':[100, 40, 75]
}).melt(id_vars=['branch'], var_name= 'Year', value_name='No_of_students')
```

```python
# Melt function in real world
death = pd.read_csv('time_series_covid19_deaths_global.csv')
confirm = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
death
```

```python
death = death.melt(id_vars=['Province/State','Country/Region','Lat','Long'], var_name='date', value_name='total_death')
death
```

```python
confirm = confirm.melt(id_vars=['Province/State','Country/Region','Lat','Long'], var_name='Date', value_name='No_of death')
confirm
```

```python
# pivot table 
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
# # findout what is the average toal bill with respect to male
df.groupby(['sex', 'smoker'])['total_bill'].mean().unstack()
```

```python
df.pivot_table(index='sex', columns='smoker', values='total_bill')
```

```python
df
```

```python
df.pivot_table(index='day', columns='sex', values='total_bill', aggfunc='sum')
```

```python
df.pivot_table(index='sex', columns='smoker', values='total_bill')
```

```python
df.pivot_table(index=['sex', 'smoker'], columns=['day', 'time'], aggfunc=({'size':'mean', 'tip':'max', 'total_bill':'sum'}))
```

```python
df.pivot_table(index='sex', columns='smoker', values='total_bill', aggfunc='sum', margins=True)
```

```python
df = pd.read_csv('expense_data.csv')
df
```

```python
import datetime as dt 
df['months'] = pd.to_datetime(df['Date']).dt.month_name()
df
```

```python
df.pivot_table(index='Category', columns='months', values='INR', aggfunc='sum').transpose().plot(kind='bar')
```

```python
df.pivot_table(index='months', columns='Income/Expense', values='INR').plot()
```
